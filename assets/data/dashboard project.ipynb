```python
bonnie_data = mega_df[['ID', 'loan_amnt', 'funded_amnt', 'int_rate', 'installment', 'grade',
       'emp_title', 'emp_length', 'home_ownership', 'annual_inc', 'issue_d',
       'loan_status', 'title', 'addr_state','dti',
        'earliest_cr_line', 'fico_range_low', 'fico_range_high',
       'open_acc','pub_rec_bankruptcies']]
bonnie_data.head(25)
```

```python
# Total number of loans per state
total_state_loans = bonnie_data.groupby('addr_state').count()
total_state_loans
```

```python
# Total sum of money funded to each state
total_money_loaned_st = bonnie_data.groupby(['addr_state']).sum()[['loan_amnt', 'funded_amnt']]
total_money_loaned_st = total_money_loaned_st.astype(int)
total_money_loaned_st

# total_money_loaned_df = pd.DataFrame(total_money_loaned_st)
# total_money_loaned_df

# # Rename column for future merge
# total_money_loaned_df = total_money_loaned_df.rename(columns = {'loan_amnt' : "sum_loaned"})
# total_money_loaned_df
```
